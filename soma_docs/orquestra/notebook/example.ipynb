# How to validate and compute rewards for list of molucule

```python
from orquestra.drug.drug.validator import  ConditionFilters
from orquestra.drug.drug.validator import GeneralFilter, PainFilter, WehiMCFilter,SybaFilter
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/mohamad/workspace/repos/orquestra-drug-discovery/notebook/data/KRAS_G12D_inhibitors.csv")
smi = df.smiles.to_list()
```

```python
filter_lists=[PainFilter(),GeneralFilter()]
weight_lists=[5.0,1.1]
```

```python
filter = ConditionFilters(filter_lists=filter_lists,weight_lists=weight_lists)
```

```python
results = filter.apply_all(smi)
```
