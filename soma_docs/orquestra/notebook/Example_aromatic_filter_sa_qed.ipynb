```python
import os
import sys
import pandas as pd
from orquestra.drug.metrics import load_aromatic_substructures,calculate_sa_score,calculate_qed_score,filter_aromatic_molecules
# Load the aromatic substructures from the text file
substructures = load_aromatic_substructures('aromatic_substructures.txt')

# Define a list of SMILES for testing
smiles_list = [
    'c1ccccc1',  # Benzene
    'c1ccncc1',  # Pyridine
    'CCO',       # Ethanol (non-aromatic)
    'c1ccc2ccccc2c1',  # Naphthalene
]

# Apply the filters and calculate scores
results = []
for smiles in smiles_list:
    contains_aromatic = filter_aromatic_molecules(smiles, substructures)
    sa_score = calculate_sa_score(smiles)
    qed_score = calculate_qed_score(smiles)
    results.append({
        'SMILES': smiles,
        'Contains Aromatic Structure': contains_aromatic,
        'SA Score': sa_score,
        'QED Score': qed_score
    })

# Display the results in a DataFrame
df = pd.DataFrame(results)
df
```
